```python
def resize(image, scale):
  width, height, channel = image.shape
  methods = [cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_CUBIC]

  width = int(scale * width)
  height = int(scale * height)
  dim = (height, width)
  des_image = []
  des_image.append(cv2.resize(image, dim, interpolation = methods[0]))
  des_image.append(cv2.resize(image, dim, interpolation = methods[1]))
  des_image.append(cv2.resize(image, dim, interpolation = methods[2]))
  return des_image
```

## Rotate (15, 45, 90)

```python
def Rotate(image, angle):
  methods = [cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_CUBIC]
  height, width = image.shape[:2]
  center = (width/2, height/2)

  rotate_matrix = cv2.getRotationMatrix2D(center, angle, 1)
 
  des_image = []
  des_image.append(cv2.warpAffine(src=image, M=rotate_matrix, dsize=(width, height), flags=methods[0]))
  des_image.append(cv2.warpAffine(src=image, M=rotate_matrix, dsize=(width, height), flags=methods[1]))
  des_image.append(cv2.warpAffine(src=image, M=rotate_matrix, dsize=(width, height), flags=methods[2]))

  return des_image


```

```python
images = [cat, wdg4, wat1str1]
resize_cat = Rotate(cat, 15)
fig = plt.figure(figsize=(10, 10))
plt.subplot(2,2,1)
plt.imshow(cat)
plt.subplot(2,2,2)
plt.imshow(resize_cat[0])
plt.subplot(2,2,3)
plt.imshow(resize_cat[1])
plt.subplot(2,2,4)
plt.imshow(resize_cat[2])
plt.show()
```

```python
images = [cat, wdg4, wat1str1]
resize_wdg4 = Rotate(wdg4, 45)
fig = plt.figure(figsize=(10, 10))
plt.subplot(2,2,1)
plt.imshow(wdg4)
plt.subplot(2,2,2)
plt.imshow(resize_wdg4[0])
plt.subplot(2,2,3)
plt.imshow(resize_wdg4[1])
plt.subplot(2,2,4)
plt.imshow(resize_wdg4[2])
plt.show()
```

## Translation

```python
def Translation(image, xTrans, yTrans):
  methods = [cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_CUBIC]
  height, width = cat.shape[:2]

  tx, ty = width / xTrans, height / yTrans
 

  translation_matrix = np.array([ [1, 0, tx],
                                  [0, 1, ty]], dtype=np.float32)
 
  des_image = []
  des_image.append(cv2.warpAffine(src=image, M=translation_matrix, dsize=(width, height), flags=methods[0]))
  des_image.append(cv2.warpAffine(src=image, M=translation_matrix, dsize=(width, height), flags=methods[1]))
  des_image.append(cv2.warpAffine(src=image, M=translation_matrix, dsize=(width, height), flags=methods[2]))

  return des_image

```

## Shear (0.5 theo x)

```python
def shear_x(image, x):
  h, w = image.shape[0:2]
  methods = [cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_CUBIC]

  input = np.array([[0, 0], [0, w], [h, w], [h, 0]], dtype='float32')
  k = int(w*0.5)
  output = np.array([[0, k], [0, w+k], [h, w], [h, 0]], dtype='float32')
  shear_matrix = cv2.getPerspectiveTransform(input,output)

  des_image = []
  des_image.append(cv2.warpPerspective(src=image, M=shear_matrix, dsize=(w, h), flags=methods[0]))
  des_image.append(cv2.warpPerspective(src=image, M=shear_matrix, dsize=(w, h), flags=methods[1]))
  des_image.append(cv2.warpPerspective(src=image, M=shear_matrix, dsize=(w, h), flags=methods[2]))
  
  return des_image
```

```python
def Shear_x(img, x):
  rows, cols = img.shape[:2]
  methods = [cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_CUBIC]
  M = np.float32([[1, x, 0],
              [0, 1  , 0]])
  nW =  img.shape[1] + abs(x*img.shape[0])
  sheared_img = cv2.warpAffine(img, M, (int(nW), img.shape[0]))

  return sheared_img


```

```python
res_cat = Shear_x(cat, 0.5)
plt.imshow(res_cat)
```

```python
res_wdg4 = Shear_x(wdg4, 0.5)
plt.imshow(res_wdg4)
```

```python
res_wat1str1 = Shear_x(wat1str1, 0.5)
plt.imshow(res_wat1str1)
```
